```python
import pandas as pd
from collections import Counter
import numpy as np

# 데이터 로드
df = pd.read_csv("merged_portfolio_summary_with_range.csv")

# 문자열 통일 (QQQ/SCHD → QQQ-SCHD)
df['Combination'] = df['Combination'].str.replace('/', '-')

# 각 조합 내 ETF 분리
df['Tickers'] = df['Combination'].apply(lambda x: x.split('-'))

# 상위 n개 조합 추출 (공격형/안정형용으로 별도 추출)
n_top = 100  # 조합 수가 적다면 줄여도 됩니다

# 🎯 공격형 조합: 높은 수익률 + 좋은 Sharpe 비율
aggressive_df = df.sort_values(
    by=['Annual Return', 'Return-to-Volatility Ratio'], ascending=[False, False]
).head(n_top)

# 🛡️ 안정형 조합: 낮은 변동성 + 낮은 Range-to-Variance
defensive_df = df.sort_values(
    by=['Annual Volatility', 'Range-to-Variance Ratio'], ascending=[True, True]
).head(n_top)

# Ticker 빈도 계산
def get_top_etfs(df_sub):
    tickers = df_sub['Tickers'].sum()
    counts = Counter(tickers)
    scores = {}

    # 점수화: 수익률 기반 가중치 적용
    for idx, row in df_sub.iterrows():
        for t in row['Tickers']:
            if t not in scores:
                scores[t] = 0
            scores[t] += row['Annual Return']  # 가중치: 수익률로 합산

    # 정렬하여 상위 4~6개 종목 선택
    top_etfs = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    top_etfs = top_etfs[:6]  # 최대 6개 선택

    return top_etfs

top_aggressive = get_top_etfs(aggressive_df)
top_defensive = get_top_etfs(defensive_df)

# 비중 계산 (정규화)
def normalize_weights(etf_scores):
    total = sum(score for _, score in etf_scores)
    return [(etf, round(score / total, 4)) for etf, score in etf_scores]

aggressive_portfolio = normalize_weights(top_aggressive)
defensive_portfolio = normalize_weights(top_defensive)

# 결과 출력
print("🎯 최종 공격형 포트폴리오 (4~6개 ETF 기준)")
for etf, weight in aggressive_portfolio:
    print(f"{etf}: {weight*100:.2f}%")

print("\n🛡️ 최종 안정형 포트폴리오 (4~6개 ETF 기준)")
for etf, weight in defensive_portfolio:
    print(f"{etf}: {weight*100:.2f}%")
```

```python
import pandas as pd
from collections import Counter
import numpy as np
import random

# 데이터 로드
df = pd.read_csv("merged_portfolio_summary_with_range.csv")

# 문자열 통일 (QQQ/SCHD → QQQ-SCHD)
df['Combination'] = df['Combination'].str.replace('/', '-')

# 각 조합 내 ETF 분리
df['Tickers'] = df['Combination'].apply(lambda x: x.split('-'))

# 상위 n개 조합 추출
n_top = 100

# 🎯 공격형 조합: 높은 수익률 + 좋은 Sharpe 비율
aggressive_df = df.sort_values(
    by=['Annual Return', 'Return-to-Volatility Ratio'], ascending=[False, False]
).head(n_top)

# 🛡️ 안정형 조합: 낮은 변동성 + 낮은 Range-to-Variance
defensive_df = df.sort_values(
    by=['Annual Volatility', 'Range-to-Variance Ratio'], ascending=[True, True]
).head(n_top)

# Ticker 빈도 및 점수 계산 함수
def get_etf_scores(df_sub, score_col='Annual Return'):
    scores = {}
    for idx, row in df_sub.iterrows():
        for t in row['Tickers']:
            scores[t] = scores.get(t, 0) + row[score_col]
    return scores

# 포트폴리오 생성 함수
def generate_portfolios(scores_dict, num_portfolios=15, min_etfs=4, max_etfs=6):
    etfs = list(scores_dict.keys())
    num_etfs = len(etfs)

    # ETF 수가 너무 적은 경우 대응
    min_k = min(min_etfs, num_etfs)
    max_k = min(max_etfs, num_etfs)

    if num_etfs < min_etfs:
        print(f"⚠️ ETF 수가 {min_etfs}개보다 적습니다. 가능한 조합으로 생성합니다.")

    portfolios = []

    for _ in range(num_portfolios):
        k = random.randint(min_k, max_k)
        selected = random.sample(etfs, k=k)
        total_score = sum(scores_dict[etf] for etf in selected)
        weights = [(etf, round(scores_dict[etf] / total_score, 4)) for etf in selected]
        portfolios.append(weights)

    return portfolios

# 공격형 & 안정형 점수 계산
aggressive_scores = get_etf_scores(aggressive_df, score_col='Annual Return')
defensive_scores = get_etf_scores(defensive_df, score_col='Return-to-Volatility Ratio')

# 포트폴리오 15개씩 생성
aggressive_portfolios = generate_portfolios(aggressive_scores, num_portfolios=15)
defensive_portfolios = generate_portfolios(defensive_scores, num_portfolios=15)

# 결과 출력
print("🎯 공격형 포트폴리오 15가지")
for i, portfolio in enumerate(aggressive_portfolios, 1):
    print(f"\n[{i}]")
    for etf, weight in portfolio:
        print(f"{etf}: {weight*100:.2f}%")

print("\n🛡️ 안정형 포트폴리오 15가지")
for i, portfolio in enumerate(defensive_portfolios, 1):
    print(f"\n[{i}]")
    for etf, weight in portfolio:
        print(f"{etf}: {weight*100:.2f}%")
```

```python
import pandas as pd
from collections import Counter
import numpy as np
import random

# 데이터 로드
df = pd.read_csv("merged_portfolio_summary_with_range.csv")

# 문자열 통일 (QQQ/SCHD → QQQ-SCHD)
df['Combination'] = df['Combination'].str.replace('/', '-')

# 각 조합 내 ETF 분리
df['Tickers'] = df['Combination'].apply(lambda x: x.split('-'))

# 상위 n개 조합 추출
n_top = 100

# 🎯 공격형 조합: 높은 수익률 + 좋은 Sharpe 비율
aggressive_df = df.sort_values(
    by=['Annual Return', 'Return-to-Volatility Ratio'], ascending=[False, False]
).head(n_top)

# 🛡️ 안정형 조합: 낮은 변동성 + 낮은 Range-to-Variance
defensive_df = df.sort_values(
    by=['Annual Volatility', 'Range-to-Variance Ratio'], ascending=[True, True]
).head(n_top)

# Ticker 빈도 및 점수 계산 함수
def get_etf_scores(df_sub, score_col='Annual Return'):
    scores = {}
    for idx, row in df_sub.iterrows():
        for t in row['Tickers']:
            scores[t] = scores.get(t, 0) + row[score_col]
    return scores

# 포트폴리오 생성 함수
def generate_portfolios(scores_dict, num_portfolios=20, min_etfs=4, max_etfs=6):
    etfs = list(scores_dict.keys())
    num_etfs = len(etfs)

    min_k = min(min_etfs, num_etfs)
    max_k = min(max_etfs, num_etfs)

    portfolios = []

    for _ in range(num_portfolios):
        k = random.randint(min_k, max_k)
        selected = random.sample(etfs, k=k)
        total_score = sum(scores_dict[etf] for etf in selected)
        weights_raw = [scores_dict[etf] / total_score for etf in selected]

        weights_rounded = [round(w, 4) for w in weights_raw]
        diff = 1 - sum(weights_rounded)
        if abs(diff) > 0:
            max_idx = weights_rounded.index(max(weights_rounded))
            weights_rounded[max_idx] += diff
            weights_rounded[max_idx] = round(weights_rounded[max_idx], 4)

        weights = list(zip(selected, weights_rounded))
        portfolios.append(weights)

    return portfolios

# 점수 계산
aggressive_scores = get_etf_scores(aggressive_df, score_col='Annual Return')
defensive_scores = get_etf_scores(defensive_df, score_col='Return-to-Volatility Ratio')

# 포트폴리오 생성 (20개씩)
aggressive_portfolios = generate_portfolios(aggressive_scores, num_portfolios=20)
defensive_portfolios = generate_portfolios(defensive_scores, num_portfolios=20)

# DataFrame 변환 함수
def portfolios_to_df(portfolios, type_label='Aggressive'):
    portfolio_rows = []
    for idx, portfolio in enumerate(portfolios, 1):
        row = {'Portfolio ID': f'{type_label}_{idx}'}
        for etf, weight in portfolio:
            row[etf] = weight
        portfolio_rows.append(row)
    df_portfolio = pd.DataFrame(portfolio_rows)
    
    # 빈칸을 0으로 채우기
    df_portfolio = df_portfolio.fillna(0)
    
    # 합계 체크 (1±0.001 허용)
    df_portfolio['Total Weight'] = df_portfolio.drop(columns=['Portfolio ID']).sum(axis=1)
    for idx, total in enumerate(df_portfolio['Total Weight'], 1):
        if not (0.999 <= total <= 1.001):
            print(f"⚠️ {type_label}_{idx} 합계가 1이 아닙니다: {total}")
    df_portfolio = df_portfolio.drop(columns=['Total Weight'])
    
    return df_portfolio

# DataFrame 생성
df_aggressive = portfolios_to_df(aggressive_portfolios, 'Aggressive')
df_defensive = portfolios_to_df(defensive_portfolios, 'Defensive')

# CSV 저장
df_aggressive.to_csv("aggressive_portfolios.csv", index=False)
df_defensive.to_csv("defensive_portfolios.csv", index=False)

print("✅ CSV 저장 완료: aggressive_portfolios.csv / defensive_portfolios.csv")
```
